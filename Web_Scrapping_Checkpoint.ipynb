```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Step 1: Get the Wikipedia page content
url = "https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 2: Locate the table
# Wikipedia tables of interest are typically labeled with the 'wikitable' class
table = soup.find_all('table', {'class': 'wikitable'})[0]  # Selecting the first table

# Step 3: Extract headers
headers = [th.text.strip() for th in table.find_all('th')]
print("Number of headers:", len(headers))
print("Headers:", headers)

# Step 4: Extract rows and handle row length inconsistencies
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    row_data = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
    # Only add rows that match the length of the headers
    if len(row_data) == len(headers):
        data.append(row_data)
    else:
        print("Row with length mismatch:", row_data)  # Optional debugging

   
# Step 5: Create DataFrame
df = pd.DataFrame(data, columns=headers)

# Display the first few rows of the DataFrame
df.head(20)
```
