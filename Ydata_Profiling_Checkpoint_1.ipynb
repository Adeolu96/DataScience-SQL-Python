```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
```

```python
#DATA EXPLORATION WITH PANDAS
#Download the provided dataset
#Use pandas to read the file into a pandas Dataframe
df = pd.read_csv('EDUCATION_ATTAINMENT.csv')
#Randomly sample 500 rows from the DataFrame for speed and to reduce file size
sampled_df = df.sample(n=500, random_state=1)
sampled_df.head(10)
```

```python
#Display general information about the dataset.
sampled_df.info()
```

```python
#Run descriptive analysis on the dataset using the 'describe' function.
sampled_df.describe()
```

```python
#DATA EXPLORATION WITH YDATA-PROFILING
#Use ydata-profiling to generate a report of the provided dataset.
profile = ProfileReport(sampled_df, title="Profiling Report", explorative=True)
```

```python
#Display the report in a Jupyter Notebook
profile.to_notebook_iframe()
```

```python
#Look for missing values in the dataset. You can use the report generated by ydata to identify missing values.
# Check for missing values in the DataFrame
missing_values = sampled_df.isnull().sum()

# Display the missing values
print(missing_values[missing_values > 0])
```

```python
#Look for correlations between different columns in the dataset. 
numeric_df = sampled_df.select_dtypes(include=['number'])
# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```
